# Grafication the dataset of the channels 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import os

carpeta = 'Data_messages'
dataframes = []
for archivo in os.listdir(carpeta):
    if archivo.endswith('.csv'):  # Verificar si el archivo es un archivo CSV
        ruta = os.path.join(carpeta, archivo)  # Obtener la ruta completa del archivo
        df = pd.read_csv(ruta)  # Leer el archivo CSV y crear un DataFrame
        dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
merged_df = pd.concat(dataframes, ignore_index=True)

# Guardar el DataFrame resultante en un archivo CSV
merged_df.to_csv('Data_messages/merged_dataset.csv', index=False)
```

```python
df = pd.read_csv('Data_messages/merged_dataset.csv')
df
```

```python
user_freq = df['user'].value_counts()

# Graficar
plt.figure(figsize=(10, 6))
user_freq.plot(kind='bar')
plt.title('Frecuencia de Mensajes por Usuario')
plt.xlabel('Usuario')
plt.ylabel('Cantidad de Mensajes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calcular la frecuencia de mensajes por usuario y seleccionar los top N usuarios
top_n_users = 10
user_freq = df['user'].value_counts().nlargest(top_n_users)

# Graficar
plt.figure(figsize=(10, 6))
user_freq.plot(kind='bar')
plt.title(f'Top {top_n_users} Usuarios con MÃ¡s Mensajes')
plt.xlabel('Usuario')
plt.ylabel('Cantidad de Mensajes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calcular la frecuencia de mensajes por usuario
user_freq = df['user'].value_counts()

# Graficar
plt.figure(figsize=(8, 8))
user_freq.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('DistribuciÃ³n de Mensajes por Usuario')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
